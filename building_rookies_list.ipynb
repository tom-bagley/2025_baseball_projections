```python
import pandas as pd
```

```python
actual = pd.read_csv('./Resources/all_data.csv')
actual = actual[['IDfg','Season', 'Name', 'H', '1B', '2B', '3B', 'HR', 'R', 'RBI', 'BB', 'SH', 'SB', 'wRC', 'WAR', 'G', 'AB', 'PA', 'AVG', 'BB%','OBP', 'SLG', 'OPS', 'ISO', 'wRAA', 'wRC+']]
actual = actual[(actual['Season'] >= 2010) & (actual['Season'] <= 2024)]
actual['IDfg'] = actual['IDfg'].astype(str)
```

```python
steamer = pd.read_csv('./Resources/steamer_preseason_projections/all_steamer_data.csv')
```

```python
steamer_top15 = (
    steamer.groupby(['Season', 'Team'], group_keys=False)
    .apply(lambda x: x.nlargest(15, 'PA'))
)
```

```python
from pybaseball.analysis.projections import MarcelProjectionsBatting
```

```python
# Assuming the output is a list of DataFrames
marcel_batting = MarcelProjectionsBatting()


```

```python
# Create an empty list to store the data for each year
projections_data = []

# Loop over each year from 2012 to 2024
for year in range(2012, 2025):
    projections = MarcelProjectionsBatting(year)
    projections['Season'] = year  # Add a 'Season' column to differentiate the years
    projections_data.append(projections)

# Combine all years' projections into a single DataFrame
all_projections = pd.concat(projections_data, ignore_index=True)

print(all_projections)
```
